```python
import pandas as pd
```

```python
df = pd.read_csv("https://opendata.ecdc.europa.eu/covid19/casedistribution/csv")
```

```python
df.head()
```

```python
dfEuropa = df[df.continentExp == "Europe"]
```

```python
dfEuropa.head()
```

```python
dfEuropa.groupby(by="countriesAndTerritories")[['cases','deaths']].sum().sort_values('cases',ascending=False)
```
