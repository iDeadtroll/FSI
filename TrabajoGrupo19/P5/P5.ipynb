```python

import pandas as pd
```

### Importamos el dataset en formato csv desde un repositorio público y mostramos el dataframe

```python
Jugadores = 'https://raw.githubusercontent.com/iDeadtroll/FSI/main/TrabajoGrupo19/P5/19_ucl_stats.csv'
df1 = pd.read_csv(Jugadores)
df1
```

## 1.2 Realizando filtrado de los datos

### Seleccion de columnas y filas específicas.

```python
media_edad_por_club = df1.groupby("Club")["Age"].mean()
media_edad_por_club
```

### Convertir la Serie a un DataFrame y restablece el índice

```python
df_media_edad = media_edad_por_club.reset_index()
```

### Renombrar las columnas y mostrar la columna Media_Edad con 2 cifras decimales

```python
df_media_edad.columns = ["Club", "Media_Edad"]
df_media_edad['Media_Edad'] = df_media_edad['Media_Edad'].round(2)
df_media_edad
```

## 1.3 Construyendo un nuevo dataframe a partir del original.

```python
df1
```

### Calculamos valores para un nuevo campo

### Asegúramos que 'MP' no contenga valores 0 para evitar la división por cero

```python
df2 = df1[df1['MP'] != 0]
```

### Creamos un dataframe 'df2' con una columna calculada a partir otras

```python
df2['(G+A)/MP'] = df2['G+A'] / df2['MP']
df2[['MP','G+A','(G+A)/MP']]
```

### DataFrame resultado de juntar df1 y df2 por sus columnas comunes. 

```python
df3 = pd.merge(df1,df2)
df3
```

### Agrupar por 'Posicion' y ordenar por goles y asistencias por partido '(G+A)/MP' en orden descendente

```python
df3.groupby('Pos')[['MP','G+A','(G+A)/MP']].sum().sort_values('(G+A)/MP',ascending=False)
```
