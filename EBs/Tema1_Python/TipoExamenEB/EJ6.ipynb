```python
import pandas as pd
```

## Se dispone de tres dataframe de Pandas, llamadas vuelos_aerolinea1, vuelos_aerolinea2  y vuelos_aerolinea3, con datos de una serie de vuelos de tres compañias aereas.

## De cada vuelo se dispone de la ciudad de origen (columna Origen, de tipo cadena, en cada dataframe), la ciudad de destino (columna Destino de tipo cadena), la duracion del vuelo (columna Duracion de tipo entero) y si el vuelo tiene escala o no (columna Escala de tipo entero, con valores 0 ó 1)

```python
# Declaramos datos para cada dataframe
data1 = {
    'Origen': ['City1', 'City2', 'City3'],
    'Destino': ['City4', 'City5', 'City6'],
    'Duracion': [120, 150, 90],
    'Escala': [0, 1, 0]
}

data2 = {
    'Origen': ['City7', 'City8', 'City9'],
    'Destino': ['City10', 'City11', 'City12'],
    'Duracion': [200, 180, 210],
    'Escala': [1, 0, 1]
}

data3 = {
    'Origen': ['City13', 'City14', 'City15'],
    'Destino': ['City16', 'City17', 'City18'],
    'Duracion': [300, 250, 270],
    'Escala': [0, 1, 1]
}

# Creamos los dataframes
vuelos_aerolinea1 = pd.DataFrame(data1)
vuelos_aerolinea2 = pd.DataFrame(data2)
vuelos_aerolinea3 = pd.DataFrame(data3)
```

## a) Unir los tres dataframes por filas en uno solo

```python
vuelos_totales = pd.concat([vuelos_aerolinea1, vuelos_aerolinea2, vuelos_aerolinea3], ignore_index=True)
vuelos_totales
```

## b) Crear una columna llamada DuracionTotal que contenga la duración de cada vuelo, considerando que ésta es 100 minutos más la duracion del vuelo original del vuelo si éste tiene escala, o la misma duracion original, en caso contrario.

```python
vuelos_totales['DuracionTotal']= vuelos_totales['Duracion'] + 100 * vuelos_totales['Escala']
vuelos_totales
```

## c) Crear una tabla agrupada que contenga unicamente los promedio de la duracion total de los vuelos agrupados por cada par (Origen,Destino) distinto.

```python
vuelos_agrupados = vuelos_totales.groupby(['Origen', 'Destino'])['DuracionTotal'].mean().drop_duplicates()
vuelos_agrupados = vuelos_agrupados.reset_index()
vuelos_agrupados
```

## d) Ordena la tabla agrupada por el campo DuracionTotal

```python
vuelos_agrupados = vuelos_agrupados.sort_values('DuracionTotal')
vuelos_agrupados
```
